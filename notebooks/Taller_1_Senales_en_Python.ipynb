## Parte 1: Se√±ales discretas y convenciones

### Contexto te√≥rico

Una **se√±al discreta** es una secuencia de valores definida sobre un √≠ndice entero $n$. Una forma muy habitual (que usaremos en todo el taller) es:

$$x[n] = A \cdot \sin\left(2\pi \frac{f}{F_s} n\right) + C$$

donde:
- $A$ es la amplitud
- $f$ es la frecuencia de la se√±al (Hz)
- $F_s$ es la frecuencia de muestreo (Hz)
- $n$ es el √≠ndice discreto (muestras, no segundos)
- $C$ es una constante (offset o componente DC). En el **Bloque 1** se toma $C = 0$ (sinusoide sin offset); m√°s adelante veremos el efecto de $C \neq 0$.

### Convenci√≥n fundamental

| S√≠mbolo | Significado | Tipo |
|---------|------------|------|
| `n` | √çndice discreto | Enteros (0, 1, 2, ...) |
| `Fs` | Frecuencia de muestreo | Hz |
| `t = n/Fs` | Tiempo f√≠sico | Segundos |

**Error com√∫n:** Confundir `n` con `t`. El √≠ndice `n` **no tiene unidades de tiempo**.

### Bloque 1 ¬∑ Primera sinusoide y par√°metros base

---

#### üìù Hip√≥tesis previa (COMPLETA ANTES de ejecutar)

**Pregunta:** Si $F_s = 50$ Hz y $T = 2$ s, ¬øcu√°ntas muestras tendr√° la se√±al?

*Tu predicci√≥n:* N = `La se√±al tendr√° 100 muestras.`

---

**Par√°metros:** (en este bloque usamos $C = 0$, es decir, sin offset)
- Amplitud: $A = 1$
- Frecuencia de la se√±al: $f = 2$ Hz
- Frecuencia de muestreo: $F_s = 50$ Hz
- Duraci√≥n: $T = 5$ s
- Constante: $C = 0$

**Tareas:**
1. Calcula el n√∫mero de muestras $N$.
2. Crea el √≠ndice discreto $n = 0, 1, 2, \ldots, N-1$.
3. Calcula el instante de tiempo f√≠sico $t$.
4. Genera la se√±al $x[n]$ con la expresi√≥n anterior (con $C = 0$).

````python
# === PAR√ÅMETROS BASE (usados en todo el taller) ===
A = 1
f = 2        # Hz
Fs = 50      # Hz
T = 5        # segundos

# === IMPLEMENTACI√ìN ===
# TODO: Calcula N, genera n, calcula t, genera x
N = int(Fs * T)  # N√∫mero total de muestras
n = np.arange(N)  # Vector de √≠ndices de muestra
t = n / Fs       # Vector de tiempo correspondiente a cada muestra
x = A * np.sin(2 * np.pi * f * t)  # Se√±al sinusoidal   
````

````python
# Ejecuta esta celda para validar tu implementaci√≥n.
# === VALIDACI√ìN ===
assert N == int(Fs * T), f"N deber√≠a ser {int(Fs*T)}, pero es {N}"
assert len(n) == N, "El √≠ndice n debe tener N elementos"
assert np.allclose(t, n / Fs), "t debe ser n/Fs"
assert len(x) == N, "La se√±al x debe tener N muestras"
assert np.isclose(np.max(np.abs(x)), A, atol=0.01), "Amplitud m√°xima deber√≠a ser A"

print(f"‚úì N = {N} muestras, duraci√≥n = {T} s, Fs = {Fs} Hz")
print(f"‚úì Se√±al generada: amplitud m√°xima = {np.max(np.abs(x)):.3f}")
````

## Parte 2: Medidas b√°sicas ‚Äî media, energ√≠a y potencia

### Contexto te√≥rico

Para una se√±al discreta de $N$ muestras:

| Medida | F√≥rmula | Interpretaci√≥n |
|--------|---------|----------------|
| **Media** | $\mu = \frac{1}{N}\sum_{n=0}^{N-1} x[n]$ | Valor promedio (componente DC) |
| **Energ√≠a** | $E = \sum_{n=0}^{N-1} \lvert x[n] \rvert^2$ | Acumulado total (crece con N) |
| **Potencia media** | $P = \frac{1}{N}\sum_{n=0}^{N-1} \lvert x[n] \rvert^2$ | Promedio de energ√≠a por muestra |

**Pregunta clave:** ¬øQu√© pasa con E y P al aumentar N para una se√±al peri√≥dica?

### Bloque 3 ¬∑ Efecto del offset en la media

---

#### üìù Hip√≥tesis previa

**Pregunta:** Si en la expresi√≥n general tomas $C = 0.5$ (en lugar de $C = 0$), es decir $x_{\text{offset}}[n] = x[n] + C$:
- Media de $x[n]$ original (con $C = 0$): esperada = 0
- Media de $x_{\text{offset}}[n] = x[n] + C$ con $C = 0.5$: esperada = 0.5

*Tu predicci√≥n:* `La media de la se√±al con offset ser√° aproximadamente 0.5, ya que el offset a√±ade un valor constante a cada muestra, elevando el promedio general de la se√±al. En contraste, la se√±al original sin offset tendr√° una media cercana a 0, ya que las oscilaciones positivas y negativas se cancelan mutuamente.`

---

````python
# === IMPLEMENTACI√ìN ===


# TODO: Crea la se√±al con offset, x_offset
offset = 0.5
x_offset = x + offset


# TODO: Calcula las medias, media_x y media_x_offset
media_x = np.mean(x)
media_x_offset = np.mean(x_offset)


print(f"Media de x[n]: {media_x:.4f}")
print(f"Media de x_offset[n]: {media_x_offset:.4f}")
print(f"Diferencia: {media_x_offset - media_x:.4f}")
````

````python
# Ejecuta esta celda para validar tu implementaci√≥n.
# === VALIDACI√ìN ===
assert np.isclose(media_x_offset - media_x, offset, atol=0.01), \
    f"La diferencia de medias deber√≠a ser {offset}"
print("‚úì El offset afecta a la media como se esperaba")
````

### Bloque 4 ¬∑ C√°lculo de energ√≠a y potencia

---

#### üìù Hip√≥tesis previa

**Pregunta:** Si la duraci√≥n de la se√±al se duplica (de 2s a 4s), ¬øc√≥mo cambian:
- La energ√≠a E: ¬øse duplica / se mantiene / otra cosa?
- La potencia P: ¬øse duplica / se mantiene / otra cosa?

*Tu predicci√≥n:* `Si la duraci√≥n de la se√±al se duplica, la energ√≠a E se duplicar√°, ya que la energ√≠a es una suma acumulativa de las muestras. Sin embargo, la potencia P se mantendr√° aproximadamente constante, ya que es un promedio de la energ√≠a por muestra, y aunque haya m√°s muestras, el valor promedio no cambia significativamente para una se√±al peri√≥dica.`

---

````python
# === IMPLEMENTACI√ìN ===
# TODO: Implementa funciones de media, energ√≠a y potencia

def calcular_media(signal):
    """Calcula la media de una se√±al, sin usar la funci√≥n np.mean"""
    total = 0.0
    count = 0
    for v in signal:
        total += float(v)
        count += 1
    return total / count if count > 0 else 0.0
    

def calcular_energia(signal):
    """Calcula la energ√≠a total de una se√±al."""
    energia = 0.0
    for v in signal:
        energia += float(v) ** 2
    return energia
    

def calcular_potencia(signal):
    """Calcula la potencia media de una se√±al."""
    energia = calcular_energia(signal)
    return energia / len(signal) if len(signal) > 0 else 0.0

# Calcula para ambas se√±ales
mu_x, E_x, P_x = calcular_media(x), calcular_energia(x), calcular_potencia(x)
mu_off, E_off, P_off = calcular_media(x_offset), calcular_energia(x_offset), calcular_potencia(x_offset)

# Muestra resultados en formato tabla
print(f"{'Medida':<15} {'x[n]':>12} {'x_offset[n]':>12}")
print("-" * 40)
print(f"{'Media':<15} {mu_x:>12.4f} {mu_off:>12.4f}")
print(f"{'Energ√≠a':<15} {E_x:>12.2f} {E_off:>12.2f}")
print(f"{'Potencia':<15} {P_x:>12.4f} {P_off:>12.4f}")
````

````python
# Ejecuta esta celda para validar tu implementaci√≥n.
# === VALIDACI√ìN ===
assert P_x > 0, "La potencia debe ser positiva"
assert E_x > 0, "La energ√≠a debe ser positiva"
assert P_off > P_x, "La potencia con offset deber√≠a ser mayor"
print("\n‚úì Medidas calculadas correctamente")
````

### Bloque 6 ¬∑ Segmentaci√≥n: extraer 1 per√≠odo y a√±adir ceros

---

#### Objetivo del bloque

Preparar una se√±al **finita** a partir de la sinusoide del Bloque 1 para poder aplicar despu√©s un retardo **sin confusiones**.

La idea es:
- Tomar un segmento que corresponda con **un per√≠odo** de la se√±al original $x[n]$.
- Construir una nueva se√±al de longitud fija con **ese segmento** y, a continuaci√≥n, **una secuencia de ceros** a la derecha de la misma longitud que el segmento extraido.

---

#### Definici√≥n

Si $N_p$ es el n√∫mero de muestras por per√≠odo:

$$x_{\text{finita}}[n] = \begin{cases}
 x[n] & 0 \le n < N_p\\
 0 & N_p \le n < 2N_p
\end{cases}$$

---

#### üìù Hip√≥tesis previa

**Pregunta:** ¬øQu√© longitud tendr√° $x_{\text{finita}}[n]$ y qu√© valores tendr√° en su segunda mitad? Tendr√° longitud `2*Np` y en su segunda mitad tendr√° valores de `0`, ya que se ha definido que para los √≠ndices n mayores o iguales a Np, la se√±al toma el valor de 0.

---

````python
# === IMPLEMENTACI√ìN ===
# Vamos a extraer 1 per√≠odo de la sinusoide del Bloque 1 y a√±adir otro per√≠odo de ceros.

# TODO: Calcula el n√∫mero de muestras por per√≠odo, Np
Np = int(Fs / f)  # N√∫mero de muestras por per√≠odo de la sinusoide de frecuencia f

# TODO: Extrae un per√≠odo de x, empezando en n=0. Rellena con ceros hasta llegar a 2Np, obteniendo, x_finita
x_periodo = x[:Np]  # Un per√≠odo de la sinusoide
x_finita = np.concatenate((x_periodo, np.zeros(Np)))  # Un per√≠odo seguido de un per√≠odo de ceros   


# √çndice para representar esta se√±al (longitud 2Np)
n_pad = np.arange(len(x_finita))

# Visualizaci√≥n
plt.figure(figsize=(12, 3))
plt.stem(n_pad, x_finita, basefmt=' ')
plt.title('Se√±al de prueba: 1 per√≠odo + 1 per√≠odo de ceros (x_pad)')
plt.xlabel('n')
plt.tight_layout()
plt.show()
````

````python
# Ejecuta esta celda para validar tu implementaci√≥n.
# === VALIDACI√ìN ===

# Np debe ser un entero positivo
assert isinstance(Np, int) and Np > 0

# La se√±al debe tener longitud 2Np: 1 per√≠odo + Np ceros
assert len(x_finita) == 2 * Np

# La primera mitad coincide con x[:Np] y la segunda mitad son ceros
assert np.allclose(x_finita[:Np], x[:Np])
assert np.allclose(x_finita[Np:], 0)

print(f"‚úì Segmentaci√≥n OK: Np={Np}, len(x_finita)={len(x_finita)} (1 per√≠odo + ceros)")
````

### Bloque 7 ¬∑ Operaciones: desplazamiento (shift) y escalado

---

#### Objetivo del bloque

Aplicar un **desplazamiento** y **escalado** sobre una se√±al finita `x_finita`.

- Deslazamiento, ser√° un retardo de $n_0=10$ muestras, de forma que $x_{retardada}[n]=x_{finita}[n-n_0]$
  - Vamos a considerar que la se√±al $x_{retardada}$ sea de la misma longitud que $x_{finita}$.
- Escalado, consistir√° en multiplicar la amplitud de la se√±al por $\alpha=2$, de forma que $x_{escalada}[n]=\alpha x_{finita}[n]$

---

#### üìù Hip√≥tesis previa

- Las primeras $n_0$ muestras tras el retardo ser√°n cero, ya que la se√±al se ha desplazado hacia la derecha y no hay valores anteriores a n=0.
- ¬øC√≥mo variar√° la potencia al desplazar la se√±al? La potencia se mantendr√° constante, ya que el desplazamiento no altera los valores de las muestras, solo cambia su posici√≥n en el tiempo. Por lo tanto, la energ√≠a total y la potencia media de la se√±al no se ver√°n afectadas por el desplazamiento.
- ¬øQu√© factor esperas en la potencia al escalar por $\alpha$? La potencia se multiplicar√° por $\alpha^2$, es decir, por 4 en este caso, ya que la potencia es proporcional al cuadrado de la amplitud. Al escalar la se√±al por un factor de 2, la potencia aumentar√° en un factor de 4.

---

````python
# === IMPLEMENTACI√ìN ===
# Generar la se√±al x_retardada a partir de x_finita construida en el Bloque anterior.

n0 = 10
alpha = 2.0

# TODO: Crea la se√±al x_retardada a partir de x_finita
x_finita = np.concatenate((np.zeros(n0), x_finita))  # Retardo de n0 muestras

# TODO: Crea la se√±ala x_escalada a partir de x_finita
x_escalada = alpha * x_finita  # Escalado por factor alpha

# Potencia de las se√±ales
P_x_finita = np.mean(x_finita**2)
P_x_retardada = np.mean(x_retardada**2)
P_x_escalada = np.mean(x_escalada**2)
print(f"Potencia de x_finita: {P_x_finita:.4f}")
print(f"Potencia de x_retardada: {P_x_retardada:.4f}")
print(f"Potencia de x_escalada: {P_x_escalada:.4f}")

# Visualizaci√≥n
fig, axes = plt.subplots(1, 3, figsize=(14, 3))

axes[0].stem(np.arange(len(x_finita)), x_finita, basefmt=' ')
axes[0].set_title('Original (x_finita)')
axes[0].set_xlabel('n')

axes[1].stem(np.arange(len(x_finita)), x_retardada, basefmt=' ', linefmt='r-', markerfmt='ro')
axes[1].set_title(f'Retardo n0={n0}')
axes[1].set_xlabel('n')

axes[2].stem(np.arange(len(x_finita)), x_escalada, basefmt=' ', linefmt='g-', markerfmt='go')
axes[2].set_title(f'Escalado Œ±={alpha}')
axes[2].set_xlabel('n')

plt.tight_layout()
plt.show()
````

````python
# Ejecuta esta celda para validar tu implementaci√≥n.
# === VALIDACI√ìN ===

# Usamos los nombres de variables x_finita, x_retardada y x_escalada 
# para mantener coherencia con las se√±ales definidas antes.

# Longitud fija
assert len(x_retardada) == len(x_finita)
assert len(x_escalada) == len(x_finita)

# Primeras n0 muestras del retardo: ceros
assert np.allclose(x_retardada[:n0], 0)

# Potencias (usamos la definici√≥n emp√≠rica P = (1/N) sum x^2)
P_x = np.mean(x_finita**2)
P_delay = np.mean(x_retardada**2)
P_scaled = np.mean(x_escalada**2)

# Si n0 < N, lo que se recorta por la derecha son ceros -> la potencia se conserva
assert np.isclose(P_delay, P_x, atol=1e-12)

# Escalado: P(alpha x) = alpha^2 P(x)
assert np.isclose(P_scaled, (alpha**2) * P_x, atol=1e-12)

print("‚úì Retardo y escalado OK")
print(f"  P(x_finita)     = {P_x:.6f}")
print(f"  P(retardada)    = {P_delay:.6f} (debe coincidir)")
print(f"  P(escalada)     = {P_scaled:.6f} (debe ser Œ±^2¬∑P)")
````
